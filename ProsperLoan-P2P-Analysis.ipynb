```python
import numpy as np
import pandas as pd
import scipy
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_details = pd.DataFrame({'hourly_income':[1000,2009,24418,444478,324235,243242,3434234,7567457,9235,238237,1312,3412],
                   'hourly_expense':[651361,217371,2746,2356,13436,5732,346346,3463,1132,23534,242235,235235],
                   'family_members_count':[3,4,2,3,1,4,5,6,3,6,3,5],
                   'house_rent':[1299,2300,3411,3422,4566,4211,4600,736,672,0,734,2374],
                   'highest_income_member':["Olivia","George","Isla","Harry","Ava","Noah","Sophia","Jacobi","Freddie","Ella","Grace","Ella"]}
                   
    )
df_details
```

# A. Display the five point summary of the data.

```python
#first quartile
print(df_details.quantile(.25),'\n')
print(df_details.quantile(.50),'\n')
print(df_details.quantile(.75),'\n')
print(df_details.max(),'\n')
print(df_details.min(),'\n')
print(df_details.describe())
```

# B. What is the mean of the hourly expense?

```python
df_mean = df_details['hourly_expense'].mean()
df_mean
```

# C. What is the median of the hourly expense?

```python
df_median = df_details['hourly_expense'].quantile(.50)
df_median
```

# D. Find the family member with maximum income and using a suitable graph.

```python
df_details.head()
```

```python
sns.barplot(data=df_details,x='family_members_count',y='hourly_income')
plt.show()
## From this graph we can tell that the family members count 6 have the maximum income.
```

# E. Calculate IQR(the difference between 75% and 25% quartile) for Hourly_Income and Hourly_Expense.

```python
q3_hour = df_details['hourly_income'].quantile(.75)
print(q3_hour)

q1_hour = df_details['hourly_income'].quantile(.25)
print(q1_hour)

iqr_hour = q3_hour - q1_hour
iqr_hour

```

```python
q3_expense = df_details['hourly_expense'].quantile(.75)
print(q3_expense)

q1_expense = df_details['hourly_expense'].quantile(.25)
print(q1_expense)

iqr_expense = q3_expense - q1_expense
iqr_expense
```
